```python
def QuickSort(arr):
    less = []
    greater = []
    print(arr)
    if not arr:
        return []
    
    if len(arr)==1: 
        return arr
    pivot = arr[0]
    for i in range(len(arr)):
        if arr[i]< pivot:
            less.append(arr[i])
        else:
            greater.append(arr[i])
    return QuickSort(less)+[pivot]+QuickSort(greater)

    
```

```python
QuickSort([3,7,1,5,3,2])
```
